```python
import wfdb
import numpy as np
import pandas as pd
#df = pd.read_csv('/content/ecg-data/23.dat')

record_path = 'ecg-data/23'
ann = wfdb.rdann(record_path, 'atr')
signals, fields = wfdb.rdsamp(record_path)

# `symbol` is the label class, `sample` is the index for the signals array
df = pd.DataFrame({'symbol': ann.symbol, 'aux': ann.aux_note, 'position': ann.sample})

fs = fields['fs']
print('ECG signal size:', signals.shape)
print('ECG sampling rate:', fs, '[Hz]')
print('Labels distribution:', df.symbol.value_counts())
```

```python
from matplotlib import pyplot as plt

# Representative sample of the raw signal ...
start = 7190174
end = 7190618
signal_part  = signals[start: end, 0]

plt.figure(figsize=[15, 4])
plt.plot(signal_part, color='dimgray')

# ... and corresponding annotations
ids = (df.position > start) & (df.position < end)
df_part = df[ids]

# Shift positions of the sample beats to match the signal fragment
annotation_xs = df_part.position - start
annotation_ys = np.ones_like(annotation_xs) * 2.1
plt.plot(annotation_xs, annotation_ys, 'o', color='teal', ms=10)
plt.grid()
plt.xlabel('Sample index')
plt.ylabel('ECG [mV]')
plt.title('ECG sample')
plt.xlim(0, len(signal_part))
plt.ylim(signal_part.min(), signal_part.max() + 1.666)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate or load your original FHR signal (replace this with your data)
# For this example, we generate a synthetic FHR signal.
np.random.seed(0)  # for reproducibility
time = df['position'].values
original_fhr = df['position'].values

# Define the number of augmented samples to generate
num_augmented_samples = 5

# Data Augmentation Techniques
augmented_fhr_signals = []

for _ in range(num_augmented_samples):
    # Time Warping (Stretch or Compress)
    time_warp_factor = np.random.uniform(0.8, 1.2)  # Adjust the range as needed
    warped_time = time * time_warp_factor
    augmented_fhr = np.interp(time, warped_time, original_fhr)

    # Adding Noise
    noise_amplitude = np.random.uniform(0, 5)  # Adjust the range as needed
    augmented_fhr += np.random.normal(0, noise_amplitude, len(time))

    # Baseline Variation (Baseline Wander)
    baseline_variation_amplitude = np.random.uniform(0, 5)  # Adjust the range as needed
    augmented_fhr += np.sin(2 * np.pi * 0.02 * time) * baseline_variation_amplitude

    augmented_fhr_signals.append(augmented_fhr)

# Visualize the original and augmented FHR signals
plt.figure(figsize=(12, 6))
plt.plot(time, original_fhr, label='Original FHR', color='b')

for i, augmented_fhr in enumerate(augmented_fhr_signals):
    plt.plot(time, augmented_fhr, label=f'Augmented FHR {i+1}', alpha=0.7)

plt.xlabel('Time (seconds)')
plt.ylabel('Fetal Heart Rate (bpm)')
plt.legend()
plt.title('Original and Augmented FHR Signals')
plt.grid()
plt.show()

```

```python
import os
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report

# Define the directory where your data files are located
#data_directory = '/content/drive/MyDrive/ctu-chb-intrapartum-cardiotocography-database-1.0 (1).0'
data_directory ='/content/ecg-data/'

# Define a function to extract features from ECG signals
def extract_features(data_file, header_file):
    try:
        # Load the ECG signal from the data file with the appropriate encoding
        ecg_signal = np.loadtxt(data_file, encoding='latin1')  # Use 'latin1' encoding for non-UTF-8 data
    except Exception as e:
        print(f"Error loading data file {data_file}: {e}")
        return None  # Return None to indicate an error

    # Parse the header file to extract relevant information
    with open(header_file, 'r', encoding='utf-8') as header:
        lines = header.readlines()

        # Initialize variables to store extracted data
        gestational_weeks = None
        maternal_age = None
        fetal_heart_rate = None  # You may need to adjust this based on the data format

        # Iterate through lines in the header file
        for line in lines:
            if line.startswith("#-- Fetus/Neonate descriptors"):
                # Extract gestational weeks
                gestational_weeks = int(lines[lines.index(line) + 1].split()[-1])
            elif line.startswith("#-- Maternal (risk-)factors"):
                # Extract maternal age
                maternal_age = int(lines[lines.index(line) + 1].split()[-1])
            elif line.startswith("#-- Signal information"):
                # Extract fetal heart rate (assuming it's present in the signal info)
                fetal_heart_rate = int(lines[lines.index(line) + 1].split()[-1])

    # Return the extracted features as a dictionary
    return {
        "GestationalWeeks": gestational_weeks,
        "MaternalAge": maternal_age,
        "FetalHeartRate": fetal_heart_rate,
        "MeanAmplitude": np.mean(ecg_signal),  # Example feature
    }

# Load and preprocess your dataset
data = []
labels = []

# Loop through the data files in the directory
for filename in os.listdir(data_directory):
    if filename.endswith(".dat") and filename[:-4] + ".hea" in os.listdir(data_directory):
        data_file = os.path.join(data_directory, filename)
        header_file = os.path.join(data_directory, filename[:-4] + ".hea")

        # Extract features and determine the label, and skip files with errors
        features = extract_features(data_file, header_file)
        if features is None:
            continue

        # Determine the label based on your criteria
        if features["FetalHeartRate"] is not None and features["FetalHeartRate"] > 120:
            label = 'risk'
        else:
            label = 'normal'

        data.append(features)
        labels.append(label)

# Convert data and labels to numpy arrays
data = np.array(data)
labels = np.array(labels)

# Check if there is any valid data
if len(data) == 0:
    print("No valid data to process.")
else:
    # Split the dataset into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.2, random_state=42)

    # Train a Random Forest Classifier
    rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
    rf_classifier.fit(X_train, y_train)

    # Predict on the test set
    rf_predictions = rf_classifier.predict(X_test)

    # Evaluate the Random Forest Classifier
    rf_accuracy = accuracy_score(y_test, rf_predictions)
    print("Random Forest Classifier Accuracy:", rf_accuracy)
    print(classification_report(y_test, rf_predictions))

    # Train an SVM Classifier
    svm_classifier = SVC(kernel='linear', C=1.0, random_state=42)
    svm_classifier.fit(X_train, y_train)

    # Predict on the test set
    svm_predictions = svm_classifier.predict(X_test)

    # Evaluate the SVM Classifier
    svm_accuracy = accuracy_score(y_test, svm_predictions)
    print("SVM Classifier Accuracy:", svm_accuracy)
    print(classification_report(y_test, svm_predictions))
```
